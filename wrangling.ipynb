# DATA WRANGLING AND VISUALIZATION
- FILTERIING - based on a condition
- Boolean Indexing - Used to filter rows


```python
import pandas as pd #loads and read files
import numpy as np #performs calculations
import matplotlib.pyplot as plt #for visualizations
import seaborn as sns #for statistical visualizations like histograms

df =pd.read_csv(r"titanic _dataset.csv")
```

```python
df.info()
```

```python
df.describe() #gives statistics of data
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median)
df.info()
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors = "coerce") #convert Age from object to numeric
df.info() 
```

```python
adults = df[df["Age"]>= 18]
adults.head()
```

```python
passengers = df[df["Pclass"]==3]
passengers.head()
```

```python
passengers["Pclass"].value_counts() #filters only passenges in class 3
```

```python
#sort fare in ascending order 
df_sorted = df.sort_values(by="Fare", ascending=True)
df_sorted.head(10)
```

#### FEATURE ENGINEERING
-Coming up with new columns from existing columns e.g age groups

```python
#create age groups using pandas
df["Age_group"] =  pd.cut(df["Age"],bins = [0,12,19,59,100], labels = ["Child", "Teen", "Adult", "Senior"])
df.head()
```

```python
#extra title from Name column
df["Title"] = df["Name"].apply(lambda name: name.split(",")[1].split(".")[0].strip())
df.head()
```

# GROUP BY AND AGGREGATION
- use groupby and agg to compute statistics for sub-groups
- Groupby, one column has to be an object, the other numerice

```python
#show how many people survived per passenger class
grouped = df.groupby("Pclass")["Survived"].agg(["count","mean"])
grouped.head()

#object column = Pclass
```

```python
df["Survived"].value_counts()
```

```python
df_survived = df[df["Survived"] == 1]
df_survived.head()
```

# Merging dataframes
 - combining two tables 
 - similar to joins

```python
df1 = pd.DataFrame({
    "ID": [1,2,3,4,5],
    "Name": ["Joy", "Elizabeth", "Christine","Hillary","Russell"]
})
df2 = pd.DataFrame({
    "ID": [1,2,3,4,5],
    "Age": [27,29,32,27,19]
})
```

```python
merged_df = pd.merge(df1,df2, on="ID")
merged_df.head()
```
